## 1. Importing Libraries

````python
import numpy as np
import pandas as pd
from pandas import DataFrame

# Libraries for retrieving data from the web and handling zip files
import requests, zipfile
from io import StringIO
import io
````

## 2. Downloading the Dataset
Download the MyAnimeList dataset from GitHub and clean the data by removing rows where `Score` is `'Unknown'`.

**Note:** Do not include this section in your `homework()` function.

````python
# Specify the url with data
url = 'https://github.com/Hernan4444/MyAnimeList-Database/archive/refs/heads/master.zip'

# Acquire data from the url
r = requests.get(url, stream=True)

# read and extract the zipfile
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()

# Load and clean the data
anime_data = pd.read_csv('MyAnimeList-Database-master/data/anime.csv')
anime_data_extracted = anime_data[anime_data['Score'] != 'Unknown'].copy()
anime_data_extracted['Score'] = pd.to_numeric(anime_data_extracted['Score'])
````

## 3. Solution
Write your solution in the `homework()` function below. The function receives `anime_data_extracted` as input and should return a `pd.DataFrame`.

````python
# Fill in the WRITE ME
# Don't forget to erase the "!!WRITE ME!!" before submitting!
def homework(anime_data_extracted):
  # 1. Estandarizar la columna 'Score'
    mean_score = anime_data_extracted['Score'].mean()
    std_score = anime_data_extracted['Score'].std()

    score_std = (anime_data_extracted['Score'] - mean_score) / std_score

    # Creamos el DataFrame inicial con la columna requerida 'Score_std'
    result = pd.DataFrame({'Score_std': score_std})

    # 2. Aplicar One-Hot Encoding a la columna 'Type' y convertir a int
    # Usamos dummy_na=False (por defecto) y extraemos las categorías específicas
    type_dummies = pd.get_dummies(anime_data_extracted['Type']).astype(int)

    # 3. Asegurar las columnas específicas en el orden solicitado
    categories = ['Movie', 'Music', 'ONA', 'OVA', 'Special', 'TV']

    # Reindexamos para garantizar que existan todas las columnas pedidas (y en ese orden)
    # Si alguna no existiera en el subset, se llenaría con 0 gracias a fill_value=0
    type_dummies = type_dummies.reindex(columns=categories, fill_value=0)

    # Combinar los DataFrames de forma horizontal
    result = pd.concat([result, type_dummies], axis=1)

    return result
````
